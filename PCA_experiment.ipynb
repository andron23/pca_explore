```python
import pickle
import numpy as np
from pathlib import Path 
import glob
import json

from sklearn.decomposition import PCA
from scipy.spatial.distance import pdist
from collections import OrderedDict
```

```python
cd "Desktop/Face classification"
```

```python
path_gen_facebank_embs = Path('gen_facebank_embs/')
with open(str(path_gen_facebank_embs)+'/Nikitina_embeddings.pickle', 'rb') as f:
    emb = pickle.load(f) 
    
with open('predicts_facebank.pickle', 'rb') as f:
    emb_facebank = pickle.load(f)    
```

```python
for e in emb:
    print(emb[e].tolist())
```

```python
embeddings = {}
names = list(emb_facebank.keys())

for name in names:
    with open(str(path_gen_facebank_embs)+f'/{name}_embeddings.pickle', 'rb') as f: current_emb = pickle.load(f) 
    for key in current_emb:
        current_emb[key] = current_emb[key].tolist()

    embeddings[name] = {
        f'{name}_main': emb_facebank[name].tolist(), 
        f'{name}_generated': current_emb
    }

#em = json.dumps(embeddings)
with open('embeddings.txt', 'w') as json_file:
  json.dump(embeddings, json_file)
```
